# **003 Creating independent and target variables**

```python
X = sonar_data.drop(60, axis = 1)
Y = sonar_data[60]
```

```python
# Now Y have text data hence we need to convert into integer data with OneHotEncoding

Y =pd.get_dummies(Y, drop_first = True)
Y.sample(10)
```

### **R ---->>>> 1**
### **M ---->>>> 0** 

```python
Y.value_counts()
```

# **004 Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2, random_state=3, stratify=Y)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

# **005 Model making with tensorflow**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30,  activation='relu'),
    keras.layers.Dense(15,  activation='relu'),
    keras.layers.Dense(1,  activation='sigmoid')
])

model.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

model.fit(X_train, Y_train, epochs = 100, batch_size = 8)
```

```python
model.evaluate(X_test,Y_test)
```

```python
# Prediction Model
Y_pred = model.predict(X_test).reshape(-1)
print(Y_pred[:10])

# Round the values to nearest integer i.e; 0 and 1
Y_pred = np.round(Y_pred)
print(Y_pred[:10])
```

```python
Y_test[:10]
```

## Print Classification Report i.e; Performance metrics

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(Y_test, Y_pred))
```

```python
# Plot confusion metrics
import seaborn as sns
cm = tf.math.confusion_matrix(labels = Y_test, predictions = Y_pred)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot = True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

## Dropuout

```python
# Drop 0.5 i.e, 50% of neurons

model_d = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,  activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,  activation='relu'),
    keras.layers.Dropout(0.3),
    keras.layers.Dense(1,  activation='sigmoid')
])

model_d.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

model_d.fit(X_train, Y_train, epochs = 100, batch_size = 8)
```

```python
model_d.evaluate(X_test,Y_test)
```

```python
# Prediction Model
Y_pred_d = model_d.predict(X_test).reshape(-1)
print(Y_pred[:10])

# Round the values to nearest integer i.e; 0 and 1
Y_pred_d = np.round(Y_pred)
print(Y_pred[:10])


```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(Y_test, Y_pred_d))
```

```python
# Plot confusion metrics
import seaborn as sns
cm = tf.math.confusion_matrix(labels = Y_test, predictions = Y_pred_d)

plt.figure(figsize=(10,8))
sns.heatmap(cm, annot = True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
